## Perform exploratory data analysis (eda)

### Subtask:
Visualize the data to identify patterns, trends, and relationships (e.g., spread of the virus over time, affected regions).


**Reasoning**:
I will start by melting the three dataframes to transform them from a wide to a long format. Then, I will convert the 'Date' column to a datetime object, and finally merge the three dataframes into a single, comprehensive dataframe. This will allow for easier time-series analysis.



```python
df_confirmed_melted = df_confirmed.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Cases')
df_deaths_melted = df_deaths.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Deaths')
df_recovered_melted = df_recovered.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Recovered')

for df in [df_confirmed_melted, df_deaths_melted, df_recovered_melted]:
    df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y')

df_merged = pd.merge(df_confirmed_melted, df_deaths_melted, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'], how='left')
df_merged = pd.merge(df_merged, df_recovered_melted, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'], how='left')

display(df_merged.head())
```

**Reasoning**:
Now that the data is merged and the 'Date' column is in the correct format, I will group the data by 'Date' and 'Country/Region' to get daily counts per country. Then I will calculate the active cases, daily new cases, daily new deaths, and daily new recovered cases per country.



```python
df_country_date = df_merged.groupby(['Date', 'Country/Region']).agg({
    'Cases': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
}).reset_index()

df_country_date['Active Cases'] = df_country_date['Cases'] - df_country_date['Deaths'] - df_country_date['Recovered']

df_country_date['Daily New Cases'] = df_country_date.groupby('Country/Region')['Cases'].diff().fillna(0)
df_country_date['Daily New Cases'] = df_country_date['Daily New Cases'].apply(lambda x: max(0, x))

df_country_date['Daily New Deaths'] = df_country_date.groupby('Country/Region')['Deaths'].diff().fillna(0)
df_country_date['Daily New Deaths'] = df_country_date['Daily New Deaths'].apply(lambda x: max(0, x))

df_country_date['Daily New Recovered'] = df_country_date.groupby('Country/Region')['Recovered'].diff().fillna(0)
df_country_date['Daily New Recovered'] = df_country_date['Daily New Recovered'].apply(lambda x: max(0, x))

display(df_country_date.head())
```

**Reasoning**:
To identify the most affected countries, I need to calculate the total confirmed, death, and recovered cases for each country across all dates. Then, I can calculate the total active cases and sort the countries by total confirmed cases.



```python
df_country_total = df_merged.groupby('Country/Region').agg({
    'Cases': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
}).reset_index()

df_country_total['Total Active Cases'] = df_country_total['Cases'] - df_country_total['Deaths'] - df_country_total['Recovered']

df_country_total_sorted = df_country_total.sort_values(by='Cases', ascending=False)

display(df_country_total_sorted.head(10))
```

**Reasoning**:
To visualize the global trends, I will group the merged dataframe by date and sum the cases, deaths, recovered, and active cases. Then I will plot the time series for each of these metrics.



```python
import matplotlib.pyplot as plt
import seaborn as sns

df_global_date = df_merged.groupby('Date').agg({
    'Cases': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
}).reset_index()

df_global_date['Active Cases'] = df_global_date['Cases'] - df_global_date['Deaths'] - df_global_date['Recovered']

df_global_date['Daily New Cases'] = df_global_date['Cases'].diff().fillna(0)
df_global_date['Daily New Cases'] = df_global_date['Daily New Cases'].apply(lambda x: max(0, x))

df_global_date['Daily New Deaths'] = df_global_date['Deaths'].diff().fillna(0)
df_global_date['Daily New Deaths'] = df_global_date['Daily New Deaths'].apply(lambda x: max(0, x))

df_global_date['Daily New Recovered'] = df_global_date['Recovered'].diff().fillna(0)
df_global_date['Daily New Recovered'] = df_global_date['Daily New Recovered'].apply(lambda x: max(0, x))

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_global_date, x='Date', y='Cases')
plt.title('Global Confirmed Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_global_date, x='Date', y='Deaths')
plt.title('Global Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_global_date, x='Date', y='Recovered')
plt.title('Global Recovered Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Recovered Cases')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_global_date, x='Date', y='Active Cases')
plt.title('Global Active Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Active Cases')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_global_date, x='Date', y='Daily New Cases')
plt.title('Global Daily New Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Daily New Cases')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_global_date, x='Date', y='Daily New Deaths')
plt.title('Global Daily New Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Daily New Deaths')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_global_date, x='Date', y='Daily New Recovered')
plt.title('Global Daily New Recovered Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Daily New Recovered Cases')
plt.show()
```

```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
import warnings
warnings.filterwarnings('ignore')

# Step 1: Download and Load Data
path = kagglehub.dataset_download("sudalairajkumar/novel-corona-virus-2019-dataset")
print("Path to dataset files:", path)

# Load the main dataset (assume it's 'covid_19_data.csv')
file_path = os.path.join(path, 'covid_19_data.csv')
df = pd.read_csv(file_path)

# Step 2: EDA
print("\n--- Dataset Info ---")
print(df.info())
print("\n--- Missing Values ---")
print(df.isnull().sum())
print("\n--- Unique Countries ---")
print(df['Country/Region'].nunique())

# Fill missing values
df['Province/State'].fillna('Unknown', inplace=True)
df['Confirmed'].fillna(0, inplace=True)
df['Deaths'].fillna(0, inplace=True)
df['Recovered'].fillna(0, inplace=True)

# Convert Date to datetime
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])

# Plot confirmed cases trend
plt.figure(figsize=(12,6))
df.groupby('ObservationDate')['Confirmed'].sum().plot()
plt.title("Global Confirmed Cases Over Time")
plt.ylabel("Total Confirmed")
plt.grid()
plt.show()

# Step 3: Feature Engineering
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
df['Day'] = df['ObservationDate'].dt.day
df['Month'] = df['ObservationDate'].dt.month

# Create label: whether the case count is above 1000 (binary classification)
df['HighConfirmed'] = (df['Confirmed'] > 1000).astype(int)

features = df[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Day', 'Month']]
target = df['HighConfirmed']

# Step 4: Preprocessing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
y = target

# Step 5: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Step 6: ML - Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("\n=== Random Forest Classifier ===")
print(classification_report(y_test, y_pred_rf))

# Step 7: DL - Neural Network
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, batch_size=128, validation_split=0.2, verbose=1)

loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"\n=== Neural Network ===\nTest Accuracy: {acc:.4f}")
```

```python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Step 1: Download and Load Dataset
path = kagglehub.dataset_download("imdevskp/corona-virus-report")
print("Path to dataset files:", path)

df = pd.read_csv(os.path.join(path, "covid_19_clean_complete.csv"))

# Step 2: EDA
print("\n--- Dataset Info ---")
print(df.info())
print("\n--- Null Values ---")
print(df.isnull().sum())
print("\n--- Sample Data ---")
print(df.head())

# Fill missing values
df['Province/State'].fillna('Unknown', inplace=True)

# Convert date to datetime and extract features
df['Date'] = pd.to_datetime(df['Date'])
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month

# Active Cases
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']

# Plot total confirmed cases over time
plt.figure(figsize=(12,5))
df.groupby('Date')['Confirmed'].sum().plot()
plt.title("Total Confirmed Cases Over Time")
plt.ylabel("Confirmed Cases")
plt.grid()
plt.show()

# Step 3: Feature Engineering
df['HighActive'] = (df['Active'] > 1000).astype(int)
df['Country'] = LabelEncoder().fit_transform(df['Country/Region'])

features = df[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Day', 'Month', 'Country']]
target = df['HighActive']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
y = target

# Step 4: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Step 5: ML - Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("\n=== Random Forest Classifier ===")
print(classification_report(y_test, y_pred_rf))

# Step 6: DL - Neural Network
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.4),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, batch_size=128, validation_split=0.2, verbose=1)

loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"\n=== Neural Network Model ===\nTest Accuracy: {acc:.4f}")
```
